<font size="6">4.Data Visualisation and Analysis<font>

1. Which team has won the toss the most number of times?

```python
tw=df['toss_winner']
plt.figure(figsize=(8, 6))
sns.countplot(tw)
plt.title('Count of Toss Wins by Each Team in IPL')
plt.xlabel('Team')
plt.ylabel('Count of Toss Wins')
plt.xticks(rotation=90)
plt.show()
```

Here,Mumbai Indians won the toss most number of times where as Rising Pune Superiants and Pune Warriors won least number of tosses.

2. What mumbai indians preferred after winning a toss?

```python
mum=df[df['toss_winner']=='Mumbai Indians']
z=mum['toss_decision'].value_counts()
plt.figure(figsize=(8, 6))
sns.countplot(x=z.index,hue=z)
plt.title('Toss Decisions by Mumbai Indians in IPL Matches')
plt.xlabel('Toss Decision')
plt.ylabel('Count')
plt.show()
```

We can see that mumbai indians mostly preferred batting after winning a toss. Both ratio are almost same.

3.What is the head-to-head winning count between Mumbai Indians and Chennai Super Kings in IPL matches?

```python
w=df[(df['team1']=='Mumbai Indians') & (df['team2']=='Chennai Super Kings')| (df['team2']=='Mumbai Indians') & (df['team1']=='Chennai Super Kings') ]
mum_win=w[w['winner']=='Mumbai Indians']
csk_win=w[w['winner']=='Chennai Super Kings']
mumbai_wins = len(mum_win)
chennai_wins = len(csk_win)
```

```python
teams = ['Mumbai Indians', 'Chennai Super Kings']
wins = [mumbai_wins,chennai_wins]
plt.figure(figsize=(8, 6))
plt.bar(teams, wins, color=['blue', 'yellow']) 
plt.xlabel('Teams')
plt.ylabel('Number of Wins')
plt.title('Head-to-Head Winning Count: Mumbai Indians vs Chennai Super Kings')
plt.show()
```

We can see that Mumbai Indians won most matches than CSK.

4. Which team won most of the matches in Mumbai?

```python
mum_win=df[df['city']=='Mumbai']['winner'].value_counts()
plt.figure(figsize=(10, 6))
sns.barplot(x=mum_win.index, y=mum_win.values, palette='Set2')
plt.title('Number of Matches Won by Teams in Mumbai')
plt.xlabel('Team')
plt.ylabel('Number of Wins')
plt.xticks(rotation=90)
plt.show()
```

Here, Mumbai Indians won most of the matches in Mumbai where as Gujarat Lions, Rising Pune Supergiants, Sunrisers Hyderabad and Kochi Tuskers Kerala won least matches in Mumbai.

5.How many times each team won the toss and won the match?

```python
z=df[df['toss_winner']==df['winner']]['winner']
team={}
for i in z:
    if i not in team:
        team[i]=1
    if i in team:
        team[i]+=1

# Extracting team names and win counts
teams = list(team.keys())
wins = list(team.values())

# Plotting the win counts for each team
plt.figure(figsize=(10, 6))
plt.bar(teams, wins, color='skyblue')
plt.xlabel('Team')
plt.ylabel('Number of Wins')
plt.title('Number of Wins for Each IPL Team')
plt.xticks(rotation=90)
plt.show()
```

We can see that Mumbai Indians won most most tosses as well as the matches followed by Kolkata Knight Riders  and Chennai Super Kings.

6.Which city hosted most number of matches?

```python
city=df['city'].value_counts()
plt.figure(figsize=[8, 8])
plt.barh(city.index, city.values) 
plt.xlabel('Number of Bookings')
plt.ylabel('City')  # Label for y-axis
plt.title('Cities Hosting the Most Number of Matches')
plt.gca().invert_yaxis()
plt.show()
```

```python
Here, we can see that Mumbai hosted most number of matches followed by Banglore, Kolkata and Delhi.
```

7.Which are the cricket venues in Mumbai where matches have been held?

```python
#find all venue of mumbai city
mum_ven=df[df['city']=='Mumbai']['venue'].value_counts()
plt.figure(figsize=(8,8))
mum_ven.plot(kind='pie', autopct='%1.1f%%',startangle=90)
plt.title('Distribution of Matches Played at Different Venues in Mumbai')
plt.ylabel('') 
plt.show()
```

In the above pie-chart, we can see that in Mumbai most of the matches are played at Wankhede Stadium i.e 67.1%, at Brabourne Stadium 12.9% and Dr DY Patil Sports Academy 20.0%.

8.What is the preferred choice after winning a toss in Mumbai?

```python
mum_win=df[df['city']=='Mumbai']['toss_decision'].value_counts()
plt.figure(figsize=(8, 6))
mum_win.plot(kind='bar', color='skyblue', alpha=0.7)
plt.xlabel('Toss Decision')
plt.ylabel('Frequency')
plt.title('Distribution of Toss Decisions in Matches Held in Mumbai')
plt.xticks(rotation=0)  
plt.grid(axis='y', linestyle='--', alpha=0.5) 
plt.show()
```

From above bar plot we can conclude that the preferred choice after winning a toss in Mumbai is fielding.

9.Who won most matches at M Chinnaswamy stadium?

```python
z=df[df['Stadium']=='M Chinnaswamy Stadium']['winner'].value_counts()
plt.figure(figsize=[8, 8])
plt.barh(z.index, z.values) 
plt.xlabel('Number of Tosses')
plt.ylabel('Teams') 
plt.title('Number of Tosses won by each Team')
plt.gca().invert_yaxis()
plt.show()
```

In the above bar graph, Royal Challenfers Banglore won the most matches in M Chinnaswamy stadium.

11.Which venue hosted most number of matches?

```python
z=df['venue'].value_counts()
plt.figure(figsize=(10, 8))
z.plot(kind='barh')
plt.xlabel('Number of Matches')
plt.ylabel('Venue')
plt.title('Number of Matches Hosted at Each Venue')
plt.gca().invert_yaxis()
plt.show()
```

12.How does the number of matches played vary across different seasons in the dataset?

```python
matches_per_year = df.groupby('season').size()
plt.figure(figsize=(10, 6))
matches_per_year.plot(kind='bar', color='skyblue')
plt.xlabel('Year (Season)')
plt.ylabel('Number of Matches')
plt.title('Number of Matches Played in Each Year')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

The above bar plot gives us insights that most matches are played in year 2013.

13.Which players have consistently stood out in IPL matches by receiving the highest number of 'Player of the Match' awards? 

```python
player_of_match_counts = df['player_of_match'].value_counts().head(10)
plt.figure(figsize=(8,6))
sns.barplot(x=player_of_match_counts.values, y=player_of_match_counts.index)
plt.title('Top Players with Most Player of the Match Awards')
plt.xlabel('Number of Player of the Match Awards')
plt.ylabel('Player Name')
plt.show()
```

CH Gayle received the highest number of 'Player of the Match' award.
